```python
#Данные взяты с сайта https://www.ncdc.noaa.gov/cdo-web/search 
(правда высылаются в.csv на почту по выбранным параметрам)
import matplotlib.pyplot as plt
import pandas as pd

# Загрузка данных
df = pd.read_csv(r'C:\Users\Eduard\Downloads\4063714.csv', parse_dates=['DATE'])

# Проверка данных
print(df.head())
print("\nИнформация о данных:")
print(df.info())

# Построение графика 
plt.figure(figsize=(12, 6))
```

```python
# Построения графика из DataFrame
plt.plot(df['DATE'], df['TAVG'], 
         label='Средняя температура', 
         color='red',
         linewidth=1)

# Добавляем среднюю линию для наглядности
mean_temp = df['TAVG'].mean()
plt.axhline(mean_temp, color='blue', linestyle='--', 
           label=f'Средняя: {mean_temp:.1f}F')

# Настройки графика
plt.title('Среднесуточная температура по данным NOAA', fontsize=16)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Температура (F)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
# Автоматический поворот дат для лучшей читаемости
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```
