```python
'''
Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.
'''
import pandas as pd

ratings=pd.read_csv(r'C:\Users\Eduard\Downloads\ml-latest-small\ml-latest-small\ratings.csv')
ratings.head()
        
        
```

```python
def class_movies(row):
    if row['rating']<= 2.0:
        return 'низкий рейтинг'
    elif 2.0 < row['rating'] < 4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
ratings['class'] = ratings.apply(class_movies, axis = 1)
ratings.head()
```

```python
'''
Задание 2
Используйте файл keywords.csv.
Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. 
Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
‘Центр’: [‘москва’, ‘тула’, ‘ярославль’],
‘Северо-Запад’: [‘петербург’, ‘псков’, ‘мурманск’],
‘Дальний Восток’: [‘владивосток’, ‘сахалин’, ‘хабаровск’]
}

Результат классификации запишите в отдельный столбец region.
'''
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

cities = []
for values in geo_data.values():
    for i in values:
        cities.append(i)

cities
```

```python
line = 'авито тула 126168'
data = line.split(' ')
def check(y):
    for i in y:
        if i in cities:
            for items in geo_data.items():
                if i in items[1]:
                    print(items[0])
check(data)
```

```python
keywords= pd.read_csv(r'C:\Users\Eduard\Downloads\ml-latest-small\ml-latest-small\keywords.csv')
keywords.head()
```

```python
def region_in_keyword(line):
    data = line['keyword'].split(' ')
    i=0
    for word in data:
        if word in cities:
            for items in geo_data.items():
                if word in items[1]:
                    i+=1
                    return items[0]
    if i==0:
        return 'undefined'
```

```python
keywords['region'] = keywords.apply(region_in_keyword, axis=1)
keywords.head(30)
```
